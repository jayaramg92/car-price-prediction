#Mapping columns

The 'fuelsystem' and 'carbody' columns can be dropped as there is not much difference in price difference in price between the different categories

```python
cols_to_drop.extend(['fuelsystem','carbody','enginetype'])
```

```python
df['drivewheel'] = df['drivewheel'].map({'fwd':1, 'rwd':0, '4wd':1})
df['enginelocation'] = df['enginelocation'].map({'front':1, 'rear':0})
df['doornumber'] = df['doornumber'].map({'two':1, 'four':0})
df['aspiration'] = df['aspiration'].map({'std':1, 'turbo':0})
df['fueltype'] = df['fueltype'].map({'diesel':1, 'gas':0})
df['cylindernumber'] = df['cylindernumber'].map({'four':0}).fillna(1)
```

```python
#Dropping columns
df.drop(cols_to_drop,axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

# Splitting into test and train data
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
#Scaling the variables
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_train[:] = scaler.fit_transform(df_train[:])

df_test[:] = scaler.transform(df_test[:])
```

```python
#Dividing into X and Y sets for the model building
y_train = df_train.pop('price')
X_train = df_train

y_test = df_test.pop('price')
X_test = df_test
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
#Dropping columns with high correlation
cols_to_drop = ['wheelbase','curbweight','carlength','enginelocation','carwidth','horsepower','citympg','stroke','doornumber','symboling','fueltype','boreratio']

X_train.drop(cols_to_drop, axis=1, inplace=True)
X_test.drop(cols_to_drop, axis=1, inplace=True)
```

# Creating a Linear model using sklearn

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
#testing the prediction
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
lm.score(X_train,y_train)
```

# Creating a Linear model using statsmodel

```python
import statsmodels.api as sm
X_train_sm = sm.add_constant(X_train)

lm_1 = sm.OLS(y_train,X_train_sm).fit()
print(lm_1.summary())
```

```python
# VIF check of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(X_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - lm.predict(X_train)), bins = 15)
fig.suptitle('Error Terms', fontsize = 20)
plt.xlabel('Errors', fontsize = 18)
```

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, lm.predict(X_test))
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16)     
```

The important features which can predict the price of a car are: aspiration, drivewheel, carheight, cylindernumber, enginesize, peakrpm. The model gives an R-squared of ~0.85 on both the train and test data set.
